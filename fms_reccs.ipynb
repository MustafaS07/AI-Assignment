## Data wrangling

### Subtask:
Merge the three dataframes (`df_recession`, `df_2year_yield`, and `df_10year_yield`) into a single comprehensive dataframe.


**Reasoning**:
Merge the three dataframes `df_recession`, `df_2year_yield`, and `df_10year_yield` into a single dataframe using inner join based on the date columns.



```python
# Merge df_recession and df_2year_yield based on the date columns.
df_merged = pd.merge(df_recession, df_2year_yield, left_on='observation_date', right_on='Date', how='inner')

# Merge the resulting dataframe with df_10year_yield.
df_merged = pd.merge(df_merged, df_10year_yield, on='Date', how='inner', suffixes=('_2year', '_10year'))
```

## Data preparation

### Subtask:
Prepare the merged dataframe `df_merged` for further analysis.


**Reasoning**:
Handle missing values, set 'Date' as the index, and ensure appropriate data types for the 'df_merged' dataframe.



```python
# Check for missing values
print("Missing values in df_merged:\n", df_merged.isnull().sum())

# Fill missing values using forward fill (you can adjust this based on the nature of your data)
df_merged.fillna(method='ffill', inplace=True)

# Set 'Date' as the index
df_merged.set_index('Date', inplace=True)

# Ensure appropriate data types for yield columns
for column in ['Price_2year', 'Price_10year']:
  df_merged[column] = pd.to_numeric(df_merged[column], errors='coerce')

# Inspect the prepared dataframe
display(df_merged.head())
display(df_merged.info())
```

## Feature engineering

### Subtask:
Calculate the yield curve spread and add it as a new column to the `df_merged` dataframe.


**Reasoning**:
Calculate the yield curve spread by subtracting the 2-year bond yield from the 10-year bond yield and add it as a new column in the `df_merged` dataframe.



```python
df_merged['yield_curve_spread'] = df_merged['Price_10year'] - df_merged['Price_2year']
```

## Data analysis
